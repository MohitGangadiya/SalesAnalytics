## Exploratory Data Analysis

### Q1) What is the Overall Sales Trend?

```python
all_data.head()
```

```python
all_data.columns
```

### Task 1.1 - Data Augmentation 

```python
# Adding amount column
# Amount = Price Each * Quantity - Discount %
all_data['amount'] = (all_data['price_each'] * all_data['quantity'] - (all_data['price_each'] * all_data['quantity']) * all_data['discount']).round(decimals=2)
all_data.head()
```

```python
# Reset dataFrame with columns in desired order
all_data = all_data[['order_id', 'order_date', 'ship_date', 'ship_mode', 'customer_name',
       'segment', 'state', 'country', 'market', 'region', 'product_id',
       'category', 'sub_category', 'product_name', 'price_each', 'quantity',
       'discount', 'amount', 'profit', 'shipping_cost', 'order_priority', 'year'
                    ]]
```

```python
all_data.head()
```

```python
# Adding year & month column - getting month_year from the order_date column in dataset
all_data['year_month'] = all_data['order_date'].apply(lambda x : x.strftime('%Y-%m'))
all_data['year_month'] 
```

```python
all_data.head()
```

### Task 1.2 - Data Aggregation and Grouping

```python
# Minimun year_month - starting month & year
all_data['year_month'].min()
```

```python
# Maximum year_month - ending month & year
all_data['year_month'].max()
```

```python
# Grouping data with year_month
# storing total amount and total quantity value grouped by months in df_trend
df_trend = all_data.groupby('year_month').sum()[['amount','quantity','profit']].reset_index()
df_trend.head()
```

```python
# Month with maximum sales amount
df_trend[df_trend['amount'] == df_trend['amount'].max()]
```

```python
# Month with maximum quantity ordered
df_trend[df_trend['quantity'] == df_trend['quantity'].max()]
```

### Task 1.3 - Data Visualization

```python
# Plotting graph of overall sales trend

plt.figure(figsize=(18,9))

plt.ticklabel_format(style='plain')

plt.title('Sales Trend', size=18)

plt.grid(color='maroon', linestyle='--', linewidth=0.75, axis='y', alpha=0.7)

plt.plot(df_trend['year_month'], df_trend['amount'], color='lightgreen', linewidth = 2,
         marker='o', markerfacecolor='green', markersize=10)

plt.xlabel('Year Month', size=15)
plt.ylabel('Sales in Rupees', size=15)


plt.yticks(size=12)
plt.xticks(rotation='vertical', size=12)

plt.show()
```

```python
# plotting graph of total quantity & sales vs month

fig, ax = plt.subplots(figsize=(18,9))

plt.xticks(rotation='vertical', size=15)
plt.yticks(size=12)

ax.ticklabel_format(style='plain')

axq = ax.twinx()
ax.bar(df_trend['year_month'], df_trend['amount'], color='wheat')
axq.plot(df_trend['year_month'], df_trend['quantity'], color='maroon')

plt.title('Total Sales Amount and Quantity Trend', size=18)

ax.set_xlabel('Year Month',size=15)

ax.set_ylabel('Sales in Rupees', color='g', size=15)
axq.set_ylabel('Quantity', color='r', size=15)

plt.show()
```

```python
# plotting graph of sales & profit vs month

fig, ax = plt.subplots(figsize=(18,9))

plt.xticks(rotation='vertical', size=12)
plt.yticks(size=15)

ax.ticklabel_format(style='plain')

axq = ax.twinx()
ax.plot(df_trend['year_month'], df_trend['amount'], color='red')
axq.plot(df_trend['year_month'], df_trend['profit'], color='green')

plt.title('Total Sales Amount and Profit Trend', size=18)

ax.set_xlabel('Year Month',size=15)

ax.set_ylabel('Sales', color='r', size=15)
axq.set_ylabel('Profit', color='g', size=15)

plt.show()
```

```python
# Plotting graph of overall sales Trend in specific year (2014)

year = df_trend['year_month'][df_trend['year_month'] >= '2014-01']
amount = df_trend['amount'][df_trend['year_month'] >= '2014-01']

plt.figure(figsize=(18,9))

plt.ticklabel_format(style='plain')

plt.title('Sales Trend in 2014', size=18)

plt.plot(year, amount, '-.', linewidth=1, color='lightblue', marker='o', markersize=10, markerfacecolor='blue')

plt.xlabel('Month', size=15)
plt.ylabel('Sales in Rupees', size=15)


plt.yticks(size=12)
plt.xticks(rotation='vertical', size=12)

plt.show()
```

#### Conclusion:

><p style='text-align: justify;'><i>Sales in starting of the year (first four months of the year, i.e., January, February, March, April) is quite down compared to the end of the year (last four months of the year, i.e., September, October, November, December). Again, it is stated that in the month of March the trend is upward, while in the month of July the trend is downward.</i></p>

**_"Overall, Sales in the company is growing."_**
